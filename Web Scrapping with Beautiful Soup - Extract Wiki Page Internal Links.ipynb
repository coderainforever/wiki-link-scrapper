## Extract Links to Other Wikipedia pages from a given Page

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
import re #iimport regular expression
```

```python
url = input('enter wiki page to scrap: ')
new_url = urlopen(url)
#find all the a tags
bsobj = soup(new_url.read())
for link in bsobj.findAll('a'): 
    if 'href' in link.attrs: 
        print(link.attrs['href'])
```

```python
#filtering wiki links only 
for link in bsobj.find('div',{'id': 'bodyContent'}).findAll('a',href = re.compile('^(/wiki/)((?!:).)*$')):
    if 'href' in link.attrs:
        print(link.attrs['href'])
```
